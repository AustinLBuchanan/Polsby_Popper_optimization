```python
from gerrychain import Graph
filepath = 'C:\\districting-data-2020\\'

# get tract-level graph
filename = state + '_tract.json'
G = Graph.from_json( filepath + filename )
for node in G.nodes:
    G.nodes[node]['TOTPOP'] = G.nodes[node]['P0010001']
```

```python
from number_of_districts import congressional_districts_2020
k = congressional_districts_2020[state]

import math
ideal_population = sum( G.nodes[i]['TOTPOP'] for i in G.nodes ) / k
L = math.ceil( ( 1 - deviation / 2 ) * ideal_population )
U = math.floor(( 1 + deviation / 2 ) * ideal_population )

import networkx as nx
DG = nx.DiGraph(G)
DG._L = L
DG._U = U
DG._k = k
DG._state = state
DG._level = level
DG._ordering = list(G.nodes)

print("Using k, L, U", k, L, U)
```

```python
labeling = { i : j for j in range(k) for i in districts[j] }

# set solver
from xprgrb import setSolver
setSolver("gurobi")

# create model object
from mip import build_base_mip
m = build_base_mip(DG)

# root constraints
import gurobipy as gp
root_constrs = m.addConstrs( gp.quicksum( m._r[i,j] for i in DG.nodes ) == 1 for j in range(DG._k) )

# callback info
from mip_contiguity import lcut_callback
m._callback = lcut_callback
m.Params.LazyConstraints = 1
m._numCallbacks = 0
m._numLazyCuts = 0
m._DG = DG

# add objective
from mip_objective import add_inverse_Polsby_Popper_objective
add_inverse_Polsby_Popper_objective(m, DG)

# add BVAP constraints
codes = ['P0030004'] # Black or African American alone 
codes += ['P0030011','P0030016','P0030017','P0030018','P0030019'] # Black or African American (among 2 races)
codes += ['P0030027','P0030028','P0030029','P0030030','P0030037','P0030038','P0030039','P0030040','P0030041','P0030042'] # 3
codes += ['P0030048','P0030049','P0030050','P0030051','P0030052','P0030053','P0030058','P0030059','P0030060','P0030061'] # 4
codes += ['P0030064','P0030065', 'P0030066','P0030067','P0030069'] # 5
codes += ['P0030071'] # 6

for i in DG.nodes:
    DG.nodes[i]['VAP'] = DG.nodes[i]['P0030001'] 
    DG.nodes[i]['MVAP'] = sum( DG.nodes[i][code] for code in codes )

for j in range(k):
    mvap = sum( DG.nodes[i]['MVAP'] for i in districts[j] )
    vap = sum( DG.nodes[i]['VAP'] for i in districts[j] )
    print("District",j,"has mvap =",mvap,"and vap =",vap,"or",mvap/vap*100)
    
# Impose mvap >= 0.5 * vap for districts 2 and 7 (i.e., 1 and 6 in python)
m.addConstr( gp.quicksum( DG.nodes[i]['MVAP'] * m._x[i,1] for i in DG.nodes ) >= 0.50 * gp.quicksum( DG.nodes[i]['VAP'] * m._x[i,1] for i in DG.nodes ) )
m.addConstr( gp.quicksum( DG.nodes[i]['MVAP'] * m._x[i,6] for i in DG.nodes ) >= 0.50 * gp.quicksum( DG.nodes[i]['VAP'] * m._x[i,6] for i in DG.nodes ) )

# apply local search
from mip_local_search import local_search
(labeling, new_obj, grb_time) = local_search(m, DG, labeling, radius=2)#, preserve_splits=True)
```
