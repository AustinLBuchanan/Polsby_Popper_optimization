```python
# primary settings
state = 'LA'
district_type = 'SH'
level = 'vtd'
deviation = 0.10  # 0.10 means 10% means +/-5%
minority = 'Black'
```

```python
filepath = 'C:\\districting-data-2020\\'
filename = state + '_' + level + '.json'

from read import read_graph_from_json
G = read_graph_from_json( filepath + filename )

# total population
for i in G.nodes:
    G.nodes[i]['TOTPOP'] = G.nodes[i]['P0010001'] 
    
print("number of nodes, edges:",G.number_of_nodes(),G.number_of_edges())
```

```python
from number_of_districts import number_of_districts

k = number_of_districts[state, district_type]
ideal_population = sum( G.nodes[i]['TOTPOP'] for i in G.nodes ) / k

import math
L = math.ceil(  ideal_population * (1-deviation/2) )
U = math.floor( ideal_population * (1+deviation/2) )

print("Using k, L, U =",k,L,U)
```

```python
# Step 1: carve Gingles districts
from carve import carving_heuristic
G._L = L
G._U = U
G._k = k
```

```python
# first, try mvap_excess_penalty=0.01 (i.e., 1 point charged for every 100 people above 50%)
districts01 = carving_heuristic(G, minority, mvap_excess_penalty=0.01, polsby_popper_cutoff=0.125)
```

```python
# second, try mvap_excess_penalty=0.001 (i.e., 1 point charged for every 1,000 people above 50%)
districts001 = carving_heuristic(G, minority, mvap_excess_penalty=0.001, polsby_popper_cutoff=0.125)
```

```python
# third, try mvap_excess_penalty=0.0001 (i.e., 1 point charged for every 10,000 people above 50%)
districts0001 = carving_heuristic(G, minority, mvap_excess_penalty=0.0001, polsby_popper_cutoff=0.125)
```

```python
gingles_districts = districts001
print(f"Finished carving off {len(gingles_districts)} Gingles districts.")
```

```python
from complete import recursive_bipartition_heuristic
districts = recursive_bipartition_heuristic(G, gingles_districts=gingles_districts)
```

```python
from cleanup import mip_local_search
districts = mip_local_search(G, districts, minority=minority)
```

```python
from export import export_to_baf
G._state = state
G._level = level
baf_filename = "gingles_LA_SH.csv"
labeling = { i : j for j in range(len(districts)) for i in districts[j] }
export_to_baf(G, labeling, baf_filename)
```
